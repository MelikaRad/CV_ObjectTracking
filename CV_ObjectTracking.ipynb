```python
# helper functions

def drawRectangle(frame, bbox):
    p1 = (int(bbox[0]), int(bbox[1]))
    p2 = (int(bbox[0] + bbox[2]), int(bbox[1] + bbox[3]))
    cv2.rectangle(frame, p1, p2, (255,0,0), 2, 1)
    
def displayRectangle(frame, bbox):
    plt.figure(figsize=(20,10))
    frameCopy = frame.copy()
    drawRectangle(frameCopy, bbox)
    frameCopy = cv2.cvtColor(frameCopy, cv2.COLOR_RGB2BGR)
    plt.imshow(frameCopy); plt.axis('off')    
```

###  Create the Tracker instance

```python
# Set up tracker
tracker_types = ['BOOSTING', 'MIL','KCF', 'CSRT', 'TLD', 'MEDIANFLOW', 'GOTURN','MOSSE']

# Change the index to change the tracker type
tracker_type = tracker_types[3]

if tracker_type == 'BOOSTING':
    tracker = cv2.legacy_TrackerBoosting.create()
elif tracker_type == 'MIL':
    tracker = cv2.TrackerMIL_create()
elif tracker_type == 'KCF':
    tracker = cv2.TrackerKCF_create()
elif tracker_type == 'CSRT':
    tracker = cv2.legacy_TrackerCSRT.create()
elif tracker_type == 'TLD':
    tracker = cv2.legacy_TrackerTLD.create()
elif tracker_type == 'MEDIANFLOW':
    tracker = cv2.legacy_TrackerMedianFlow.create()
elif tracker_type == 'GOTURN':
    tracker = cv2.TrackerGOTURN_create()   
else:
    tracker = cv2.legacy_TrackerMOSSE.create()
```

###  Object detection with tensorflow

```python
# detecting the objects in the first frame of video slow_traffic_small.mp4


import cv2
import matplotlib.pyplot as plt

model = cv2.dnn.readNetFromTensorflow('models/frozen_inference_graph.pb', 'models/ssd_mobilenet_v2_coco_2018_03_29.pbtxt')

# load the COCO class names
with open('models/object_detection_classes_coco.txt', 'r') as f:
    class_names = f.read().split('\n')

    
cap = cv2.VideoCapture('videos/slow_traffic_small.mp4')   

_, original_image = cap.read()
image = original_image.copy()
image_height, image_width, _ = image.shape

model.setInput(cv2.dnn.blobFromImage(image, size=(300, 300), swapRB=True, crop=False))
output = model.forward()

for detection in output[0,0,:,:]:
    score = float(detection[2])
    if score > 0.3:
        class_id = detection[1]
        class_name = class_names[int(class_id)-1]

        left = detection[3] * image_width
        top = detection[4] * image_height
        right = detection[5] * image_width
        bottom = detection[6] * image_height
        
        cv2.rectangle(image, (int(left), int(top)), (int(right), int(bottom)), (0, 0, 255), thickness=2)
        cv2.putText(image, class_name, (int(left), int(top - 5)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# plt.figure(figsize=(8,8))
# plt.subplot(211); plt.imshow(original_image[...,::-1]); plt.title('original image');
# plt.subplot(212); plt.imshow(image[...,::-1]); plt.title('result of detection');

plt.figure(figsize=(8,8))
plt.imshow(image[...,::-1]); plt.title('result of detection');
```

###  Object tracking

run one of the cells 1 or 2 or 3  
not one afer the other  
otherwise you get bounding box issues   
will be addressed later  

#### Cell 1

```python
cap = cv2.VideoCapture('videos/slow_traffic_small.mp4')   
_, original_image = cap.read()

# initial bounding box was:
bbox = (285, 179, 131, 94)

#initializing the tracker
ok = tracker.init(original_image, bbox)

while True:
    ret, frame = cap.read()
    if not ret:
        break 
    
    # Start timer
    tm = cv2.TickMeter()

    tm.start()
    ok, bbox = tracker.update(frame)
    tm.stop()
    # Draw bounding box
    if ok:
        drawRectangle(frame , bbox)
    else :
        cv2.putText(frame , "Tracking failure detected", (80,140), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)

    # Display Info
    cv2.putText(frame , tracker_type + " Tracker", (20,40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
    cv2.putText(frame , "FPS : " + str(tm.getFPS()), (20,70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
    cv2.putText(frame, class_name, (int(bbox[0]-10), int(bbox[1] - 15)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

    cv2.imshow('tracking', frame)
    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    
    
cap.release()
cv2.destroyAllWindows()
```

###  Object tracking considering if the object goes out of the frame and never comes back

one idea could be checking whether still there is an object in the bounding box or not.  
if not, puttext "object out"

#### Cell 2

```python
cap = cv2.VideoCapture('videos/slow_traffic_small.mp4')   
_, original_image = cap.read()

# initial bounding box was:
bbox = (285, 179, 131, 94)

#initializing the tracker
ok = tracker.init(original_image, bbox)

# defining a boolean variable
object_out = False

while True:
    ret, frame = cap.read()
    if not ret:
        break 
    
    # Start timer
    tm = cv2.TickMeter()
    
    if object_out:
        cv2.putText(frame , "object out", (20,90), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        cv2.imshow('tracking',frame)
    else:
        # Start timer
        tm = cv2.TickMeter()
        tm.start()
        ok, bbox = tracker.update(frame )
        tm.stop()
    
        # Display Info
        cv2.putText(frame , tracker_type + " Tracker", (20,40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
        cv2.putText(frame , "FPS : " + str(tm.getFPS()), (20,70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)


        if ok:
            # note that bbox coords are (int(left)-10, int(top)-10, int(width)+30, int(height)+20) so...
            cropped = frame[int(bbox[1]):int(bbox[1]+bbox[3]), int(bbox[0]):int(bbox[0]+bbox[2])]
            model.setInput(cv2.dnn.blobFromImage(cropped, size=(300, 300), swapRB=True, crop=False))
            output = model.forward()
            detection = output[0,0,0,:]
            score = float(detection[2])
            if score > 0.3:
                drawRectangle(frame , bbox)
                cv2.putText(frame, class_name, (int(bbox[0]-10), int(bbox[1] - 15)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

            else:
                cv2.putText(frame , "object out", (20,90), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
                cv2.imshow('tracking', frame)
                object_out = True
        else :
            cv2.putText(frame , "Tracking failure detected", (80,140), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)

        cv2.imshow('tracking', frame)
        cv2.imshow('cropped', cropped)
    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    
    
cap.release()
cv2.destroyAllWindows()
```

#### what if the object comes back?

if the object comes back, in each iteration we can check the entire frame again and detect the existing objects,  
and try to recognize it by some approach (like comparing the detected objects with the first cropped frame of the object) 

#### Cell 3

```python
cap = cv2.VideoCapture('videos/slow_traffic_small.mp4')   
_, original_image = cap.read()

#initializing the tracker
ok = tracker.init(original_image, bbox)

# initial bounding box was:
bbox = (285, 179, 131, 94)

# defining a boolean variable
object_out = False

while True:
    ret, frame = cap.read()
    if not ret:
        break 
    
    # Start timer
    tm = cv2.TickMeter()
    
    if object_out:
        cv2.putText(frame , "object out needs to be recognized again", (20,90), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
        model.setInput(cv2.dnn.blobFromImage(frame, size=(300, 300), swapRB=True, crop=False))
        output = model.forward()
        for detection in output[0,0,:,:]:
            score = float(detection[2])
            if score > 0.3:
                class_id = detection[1]
                class_name = class_names[int(class_id)-1]

                
                left = detection[3] * image_width
                top = detection[4] * image_height
                right = detection[5] * image_width
                bottom = detection[6] * image_height

                cv2.rectangle(frame, (int(left), int(top)), (int(right), int(bottom)), (0, 0, 255), thickness=2)
                cv2.putText(frame, class_name, (int(left-10), int(top - 15)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                
                
        #------------------------------------------
        # cropping each detected object
        # applying some recognition approach here
        # and setting object-out to False
        # and then start tracking again
        #------------------------------------------

        cv2.imshow('tracking',frame)
        
    else:
        # Start timer
        tm = cv2.TickMeter()
        tm.start()
        ok, bbox = tracker.update(frame )
        tm.stop()
    
        # Display Info
        cv2.putText(frame , tracker_type + " Tracker", (20,40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
        cv2.putText(frame , "FPS : " + str(tm.getFPS()), (20,70), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)


        if ok:
            # note that bbox coords are (int(left)-10, int(top)-10, int(width)+30, int(height)+20) so...
            cropped = frame[int(bbox[1]):int(bbox[1]+bbox[3]), int(bbox[0]):int(bbox[0]+bbox[2])]
            model.setInput(cv2.dnn.blobFromImage(cropped, size=(300, 300), swapRB=True, crop=False))
            output = model.forward()
            detection = output[0,0,0,:]
            score = float(detection[2])
            if score > 0.3:
                drawRectangle(frame , bbox)
                cv2.putText(frame, class_name, (int(bbox[0]-10), int(bbox[1] - 15)), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

            else:
                cv2.putText(frame , "object out", (20,90), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
                cv2.imshow('tracking', frame)
                object_out = True
        else :
            cv2.putText(frame , "Tracking failure detected", (80,140), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)

        cv2.imshow('tracking', frame)
        cv2.imshow('cropped', cropped)
    
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    
    
cap.release()
cv2.destroyAllWindows()
```
